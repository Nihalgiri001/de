```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("auto-mpg.csv")
print(df.head())
print(df.isnull().sum())

df.fillna(df.mean(numeric_only=True), inplace=True)

num_df = df.select_dtypes(include='number')
lower = num_df.quantile(0.01)
upper = num_df.quantile(0.99)
df_percentile = df[
    ~((num_df < lower) | (num_df > upper)).any(axis=1)
]
print(df_percentile)

lower = num_df.quantile(0.25)
upper = num_df.quantile(0.75)
df_IQR = df[
    ~((num_df < lower) | (num_df > upper)).any(axis=1)
]
print(df_IQR)


z = np.abs(stats.zscore(num_df))
df_zscore = df[(z < 3).all(axis=1)]
print(df_zscore)


mean = num_df.mean()
std = num_df.std()

df_std = df[
    ~((num_df < (mean - 3*std)) | (num_df > (mean + 3*std))).any(axis=1)
]
print(df_std)





df = pd.read_csv("auto-mpg.csv") 
print(df) 
df['acceleration'].fillna(df['acceleration'].median(), inplace=True) 
print(df)





df_capped = df.copy()

df_capped[num_df.columns] = df_capped[num_df.columns].clip(
    lower=num_df.quantile(0.05),
    upper=num_df.quantile(0.95),
    axis=1
)

print(df_capped)





df=pd.read_csv("iris.csv")
print("\nMean PetalWidthCm:", df['PetalWidthCm'].mean())
print("Median PetalWidthCm:", df['PetalWidthCm'].median())
print("Mode PetalWidthCm:")
print(df['Species'].mode())


print("\nRange:", df['PetalWidthCm'].max() - df['PetalWidthCm'].min())
print("Variance:", df['PetalWidthCm'].var())
print("Standard Deviation:", df['PetalWidthCm'].std())
```
