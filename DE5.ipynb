```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Load Auto MPG dataset
df = pd.read_csv("auto-mpg.csv")
print(df.head())
df.isnull().sum()

df.fillna(df.mean(numeric_only=True), inplace=True)

#Using percentile
num_df = df.select_dtypes(include='number')
lower = num_df.quantile(0.01)
upper = num_df.quantile(0.99)
df_percentile = df[
    ~((num_df < lower) | (num_df > upper)).any(axis=1)
]
print(df_percentile)

#Using IQR
lower = num_df.quantile(0.25)
upper = num_df.quantile(0.75)
df_IQR = df[
    ~((num_df < lower) | (num_df > upper)).any(axis=1)
]
print(df_IQR)


#Using Z score
z = np.abs(stats.zscore(num_df))
df_zscore = df[(z < 3).all(axis=1)]
print(df_zscore)


#Using Standard deviation
mean = num_df.mean()
std = num_df.std()

df_std = df[
    ~((num_df < (mean - 3*std)) | (num_df > (mean + 3*std))).any(axis=1)
]



# Replace missing horsepower with median
df['acceleration'].fillna(df['acceleration'].median(), inplace=True)

# Replace missing weight with mean
df['weight'].fillna(df['weight'].mean(), inplace=True)


df_capped = df.copy()

df_capped[num_df.columns] = df_capped[num_df.columns].clip(
    lower=num_df.quantile(0.05),
    upper=num_df.quantile(0.95),
    axis=1
)

print(df_capped)




# -------------------------------
# 1. Measure of Central Tendency
# -------------------------------

print("\nMean MPG:", df['mpg'].mean())
print("Median MPG:", df['mpg'].median())
print("Mode MPG:")
print(df['mpg'].mode())

# -------------------------------
# 2. Measures of Data Spread
# -------------------------------

print("\nRange:", df['mpg'].max() - df['mpg'].min())
print("Variance:", df['mpg'].var())
print("Standard Deviation:", df['mpg'].std())
```
