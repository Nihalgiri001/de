```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/iris/iris.csv")

# Box plots for all predictors
sns.boxplot(x='class', y='sepal length', data=df)
plt.show()

sns.boxplot(x='class', y='sepal width', data=df)
plt.show()

sns.boxplot(x='class', y='petal length', data=df)
plt.show()

sns.boxplot(x='class', y='petal width', data=df)
plt.show()

# Box plot for specific feature
sns.boxplot(x='class', y='sepal length', data=df)
plt.show()

# Histogram
sns.histplot(df['sepal length'])
plt.show()

# Scatter plot
sns.scatterplot(x='petal length', y='sepal length', data=df)
plt.show()




HIVE
create database shopping;

use shopping;

create table shopping1(
	code int,
	Item_name string,
	Category string,
	Place string
) row format delimited fields terminated by ‘,’;

LOAD DATA LOCAL INPATH ‘<path of a text file with data>’
OVERWRITE INTO TABLE shopping1;

create table shopping3(
				code int, 
				item_name string,
				place string
			)
			partitioned by (category string)
			clustered by (place string) into 3 buckets
			row format delimited fields terminated by ‘,’
			stored as text file;

from shopping1 txn INSERT OVERWRITE TABLE shopping3
			partition (category)
			Select txn.code, txn.item_name, txn.place, txn.category
			Distribute by category;

hadoop fs -ls /user/hive/warehouse/shopping.db/shopping3   

hadoop fs -ls /user/hive/warehouse/shopping.db/shopping3/category==utensils

hadoop fs -ls /user/hive/warehouse/shopping.db/shopping3/category==utensils/0000_1
```
