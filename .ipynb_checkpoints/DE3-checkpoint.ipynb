```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/iris/iris.csv")

# Box plots for all predictors
sns.boxplot(x='class', y='sepal length', data=df)
plt.show()

sns.boxplot(x='class', y='sepal width', data=df)
plt.show()

sns.boxplot(x='class', y='petal length', data=df)
plt.show()

sns.boxplot(x='class', y='petal width', data=df)
plt.show()

# Box plot for specific feature
sns.boxplot(x='class', y='sepal length', data=df)
plt.show()

# Histogram
sns.histplot(df['sepal length'])
plt.show()

# Scatter plot
sns.scatterplot(x='petal length', y='sepal length', data=df)
plt.show()




from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2

iris = load_iris()
X = iris.data
y = iris.target

selector = SelectKBest(score_func=chi2, k=2)
X_new = selector.fit_transform(X, y)

print("Selected Features:")
for i in selector.get_support(indices=True):
    print(iris.feature_names[i])

print("New Shape:", X_new.shape)
```
