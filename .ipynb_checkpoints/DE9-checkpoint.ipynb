```python
# LDA for Apple vs Orange Classification

import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt

# Given data
data = {
    'Weight': [150, 160, 130, 140],
    'ColorIntensity': [7, 6, 4, 3],
    'Fruit': ['Apple', 'Apple', 'Orange', 'Orange']
}

# Create DataFrame
df = pd.DataFrame(data)

# Features and target
X = df[['Weight', 'ColorIntensity']]
y = df['Fruit']

# Apply LDA
lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X, y)

# Best linear discriminant equation
print("Best Linear Discriminant:")
print("w =", lda.coef_)
print("Intercept =", lda.intercept_)
```
