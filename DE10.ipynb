```python
use booksdb
db.createCollection("books")

db.books.insertOne({title: "Harry Potter", author: "JK Rowling", price: 500})
db.books.find().pretty()

db.books.insertMany([
	{title: "Wings of fire", author: "APJ Abdul Kalam", price:250},
	{title: "1984", author: "George Orwell", price:300}
])
db.books.find()

db.books.find({price: {$eq: 300}})

db.books.updateOne(
	{title: "Harry Potter"},
	{$set: {price: 400}}
)
db.books.find().pretty()

db.books.updateMany(
	{price: {$lt: 400}},
	{$set: {price: 350}}
)
db.books.find().pretty()

db.books.deleteOne({title: "Harry Potter"})
db.books.find().pretty()

db.books.deleteMany({price: {$gt: 300}})
db.books.find().pretty()
```
